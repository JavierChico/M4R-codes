```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize
```

```python
Q = 0
alpha = 0.7; V = 0
def RHS(S):
    return -2*S**2*(alpha**2-S**2+2*S**2*np.log(S/alpha))+(alpha**2-S**2)**2-8*V*S**2
```

```python
Q = np.linspace(-6, 6, 1000)
S_vec = np.zeros_like(Q)

for j, q in enumerate(Q):
    def RHS(S):
        ans = -2*S**2*(alpha**2-S**2+2*S**2*np.log(S/alpha))+(alpha**2-S**2)**2-8*V*S**2-q
        return ans
    
    S_vec[j] = scipy.optimize.root_scalar(RHS, x0=0.5,x1=1.1, method='secant').root
    
    print()
    print(S_vec[j])
    print()
    print(q)
```

```python
plt.plot(Q,S_vec)
```
