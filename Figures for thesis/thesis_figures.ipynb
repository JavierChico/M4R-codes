```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epsilon = .1
```

```python
k = np.linspace(0,1/epsilon, 100)
plt.figure(dpi=300)
plt.plot(epsilon*k, (1-k**2*epsilon**2)/2, color = 'black', label ='Neutral curve')
plt.ylim([0,0.8])
plt.text(0.2, 0.2, 'Unstable', fontsize = 20)
plt.text(0.7, 0.5, 'Stable', fontsize = 20)
plt.legend(); plt.grid()
plt.ylabel('Magnetic number, Ma')
plt.xlabel(r'Normalised wavenumber, $\frac{k}{\epsilon}$')
plt.title('Neutral Stability curves for the long wave limit')
plt.xlim([0,1])
plt.savefig(r'neutral_stability_long_wave.pdf')
```

```python
alpha = np.linspace(0.001,1, 1000)
plt.figure(dpi=300)
plt.plot(alpha, (alpha**2-1)**2+2*(2*np.log(alpha)+1-alpha**2), color='black')
plt.grid()
plt.xlim([0,1])
plt.title(r'$f(\alpha)=(\alpha^2-1)^2+2(2\log(\alpha)+1-\alpha^2)$')
plt.xlabel(r'$\alpha$')
plt.savefig('plot_of_growth_rate_modulation_function.pdf')

plt.show()
```

```python
# dispersion relations
k = np.linspace(0,1/epsilon, 100)
D = 1
alpha = 0.5
for alpha in [0.1, 0.5, 0.9]:
    fig,ax=plt.subplots(nrows=1, ncols = 3, dpi=300, figsize = (10,3))
    for j, Ma in enumerate([0,0.25,0.5]):
        D = 1-2*Ma
        lambda_vect = k**2/16*(epsilon**2*k**2-D)*((alpha**2-1)**2-2*(alpha**2-1-2*np.log(alpha)) )- 1j/2*k*(2*np.log(alpha)+1-alpha**2)
        lambda_vect_thin = k**2/3*(D-epsilon**2*k**2)*(1-alpha)**3 + 1j*(alpha-1)**2*k
        ax[j].plot(k*epsilon, lambda_vect.real, color='black', label = 'Real')
        ax[j].plot(k*epsilon, lambda_vect.imag, color='blue', label = 'Imag')
        ax[j].plot(k*epsilon, lambda_vect_thin.real,'--', color='black', label = 'Real')
        ax[j].plot(k*epsilon, lambda_vect_thin.imag,'--', color='blue', label = 'Imag')
        ax[j].set_title(fr'$\alpha$={alpha},  Ma = {Ma}'); ax[j].grid()
        ax[j].set_ylabel(r'Growth rate, $\lambda=\lambda_r+i\lambda_i$')
        ax[j].set_xlabel(r'Normalised wavenumber, $\frac{k}{\epsilon}$')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'longwave_growth_rate_alpha_{alpha}.pdf')
    plt.show()
```
