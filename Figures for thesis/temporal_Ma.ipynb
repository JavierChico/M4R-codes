```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta=np.linspace(0,2, 200)
plt.figure(dpi=300)
plt.plot(beta, 1/(2+beta**2), color='black', label='Critical value for oscillatory case')
plt.plot(beta, np.ones_like(beta)*1/2, '--', color='blue', label='Critical value for constant current')
plt.xlabel(r'Current modulation, $\beta$')
plt.ylabel(r'Critical magnetic Bond number, $\operatorname{Ma}^*$')
plt.ylim([0,0.8])
plt.grid()
plt.title('Comparison between critical values for the magnetic Bond number')
plt.legend();plt.tight_layout()
plt.savefig('temporal_comparison_crit_Ma.pdf')
plt.show()
```

```python
t = np.linspace(0,20,1000)
A = -.1; B=1.5; C=2; Omega=3
lambda_t = A*t+B/Omega*np.sin(Omega*t)+C/(2*Omega)*np.sin(2*Omega*t)
fig = plt.figure(dpi=300)
plt.plot(t,np.exp(lambda_t), color = 'black')
plt.title(fr'A = {A}, B = {B}, C = {C} and $\Omega$ = {Omega}')
fig.suptitle(r'Plot of $S(t)=e^{\lambda(t)}$')
plt.xlabel(r'Time, $t$')
plt.ylabel(r'Amplitude, $S(t)$')
plt.grid()
plt.savefig('temporal_Ma_growth_Rate_plot.pdf')
```
