```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = np.linspace(0, 4,100)
alpha = 0.7
Ma = .01
epsilon = 0.3
def disp_relation_real(k,Ma, alpha, epsilon):
    f = 4*np.log(alpha)+(1-alpha**2)**2+2*(1-alpha**2)
    return k**2/16*(epsilon**2*k**2-1+2*Ma)*f
```

```python
plt.plot(k, disp_relation_real(k, Ma, alpha, epsilon))
plt.grid()
```
