```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import scipy
```

```python
x = np.linspace(0, 2*np.pi, 100)
t = np.linspace(0, 1, 100)
Gamma = 6
Omega = 4*np.pi
X,T = np.meshgrid(x,t)
u = np.exp(-Gamma/Omega*np.sin(Omega*T))*np.sin(X)
u_approx = (1-Gamma/Omega*np.sin(Omega*T))*np.sin(X)
u_approx2 = (1-Gamma/Omega*np.sin(Omega*T)+(Gamma/Omega*np.sin(Omega*T))**2/2)*np.sin(X)
```

```python
plt.figure(figsize=(10,10))
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))
surf = ax.plot_surface(X,T,u,cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.show()
```

```python
plt.figure(figsize=(10,10))
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))
surf = ax.plot_surface(X,T,abs(u-u_approx),cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.show()
```

```python
plt.figure(figsize=(10,10))
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))
surf = ax.plot_surface(X,T,abs(u-u_approx2),cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.show()
```
